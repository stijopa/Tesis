```python
import matplotlib.pyplot as plt
import seaborn as sns

# Variable objetivo
target = 'Precio promedio por kilogramo*_MEAN'  # Reemplaza con el nombre de tu variable objetivo

# Filtrar solo las columnas numricas, excluyendo la variable objetivo
numerical_columns = df_Pro.select_dtypes(include=['float64', 'int64']).columns.drop(target)

# Nmero de columnas por figura
cols = 2
rows = 2  # Nmero de filas por figura
fig_index = 1  # Contador para las figuras

# Crear un diagrama de dispersin para cada variable numrica contra la variable objetivo
for i, column in enumerate(numerical_columns):
    if i % (rows * cols) == 0:  # Crear una nueva figura cada 4 grficos (2 filas x 2 columnas)
        if i > 0:
            plt.tight_layout()
            plt.show()
        plt.figure(figsize=(10, 8))
    
    plt.subplot(rows, cols, (i % (rows * cols)) + 1)  # Posicin en la cuadrcula actual
    sns.scatterplot(data=df_Pro, x=column, y=target, alpha=0.6, color='skyblue')
    plt.title(f'{column} vs. {target}')
    plt.xlabel(column)
    plt.ylabel(target)

plt.tight_layout()
plt.show()
```
